# Problem 1

```python
A1 = A2 = 0.1
B1 = B2 = 0.0000008
C1 = 0.000001
C2 = 0.0000001
```

```python
alpha1 = alpha2 = 100000
```

```python
a = 0
b = 10
```

```python
def f1(t,N1, N2):
    return N1 * (A1 - B1*N1 - C1*N2)
```

```python
def f2(t,N1, N2):
    return N2 * (A2 - B2*N2 - C2*N1)
```

```python
def rk4(a, b, N, alpha1, alpha2):
    h = (b-a)/N
    t = a
    w1 = alpha1
    w2 = alpha2
    ret = np.array([[t, w1, w2]])
    for i in range(1,N+1):
        K11 = h * f1(t,w1, w2)
        K12 = h * f2(t,w1, w2)
        K21 = h * f1(t + h/2, w1 + K11/2, w2 + K12/2)
        K22 = h * f2(t + h/2, w1 + K11/2, w2 + K12/2)
        K31 = h * f1(t + h/2, w1 + K21/2, w2 + K22/2)
        K32 = h * f2(t + h/2, w1 + K21/2, w2 + K22/2)
        K41 = h * f1(t + h, w1 + K31, w2 + K32)
        K42 = h * f2(t + h, w1 + K31, w2 + K32)
        w1 = w1 + (K11 + 2*K21 + 2*K31 + K41)/6
        w2 = w2 + (K12 + 2*K22 + 2*K32 + K42)/6
        t = a + i*h
        ret = np.vstack((ret, np.array([t, w1, w2])))
    return ret
```

```python
xy = rk4(a,b,16,alpha1,alpha2)
```

```python
xy[-1,:]
```

```python
plt.plot(xy[:,0], xy[:,1],label='N1')
plt.plot(xy[:,0], xy[:,2],label='N2')
plt.legend(loc='lower left')
plt.xlabel('Time (in years)')
plt.ylabel('Population')
plt.savefig('hw6_1.pdf', bbox_inches='tight')
```

```python
exact = rk4(a,b,16,alpha1,alpha2)[-1,:][1:]
```

```python
N = [2**i for i in range(3,-1,-1)]
```

```python
N
```

```python
answers = [rk4(a,b,i,alpha1,alpha2)[-1,:][1:] for i in N]
```

```python
answers
```

```python
rel_errs_1 = [abs(i[0] - exact[0])/exact[0] for i in answers]
```

```python
rel_errs_2 = [abs(i[1] - exact[1])/exact[1] for i in answers]
```

```python
h_inverse = [i/(b-a) for i in N]
```

```python
plt.loglog(h_inverse, rel_errs_1, label='N1')
plt.loglog(h_inverse, rel_errs_2, label='N2')
plt.xlabel(r'$h^{-1}$')
plt.ylabel(r'$\epsilon_{rel}$')
plt.legend()
plt.savefig('hw6_2.pdf', bbox_inches='tight')
```

```python
plt.loglog(h_inverse, rel_errs_1, label='N1')
plt.loglog(h_inverse, rel_errs_2, label='N2')
plt.axis('equal')
plt.xlabel(r'$h^{-1}$')
plt.ylabel(r'$\epsilon_{rel}$')
plt.legend()
plt.savefig('hw6_3.pdf', bbox_inches='tight')
```

# Problem 2

```python
def f(y):
    return -3*y
```

```python
def f_an(t):
    return 50 * np.exp(-3*t)
```

```python
y0 = 50
a = 0
b = 10
h = 1
```

```python
def pred_corr(y0=y0, a=a, b=b, h=h):
    N = int((b-a)/h)
    t = np.zeros((N+1))
    t[0] = a
    w = np.zeros((N+1))
    w[0] = y0
    
    for i in range(3):
        K1 = h * f(w[i])
        K2 = h * f(w[i] + K1/2)
        K3 = h * f(w[i] + K2/2)
        K4 = h * f(w[i] + K3)
        
        w[i+1] = w[i] + (K1 + 2*K2 + 2*K3 + K4)/6
        t[i+1] = a + (i+1)*h
        
    for i in range(4, N+1):
        t[i] = a + i*h
        w[i] = w[3] + h/2 * (3*f(w[3]) - f(w[2]))
        w[i] = w[3] + (h/12) * (5*f(w[i]) + 8*f(w[3]) - f(w[2]))
        
        for j in range(3):
            t[j] = t[j+1]
            w[j] = w[j+1]
            
        t[3] = t[i]
        w[3] = w[i]
    
    w = w[4:]
    t = t[4:]
    
    return (t,w)
```

```python
(x,y) = pred_corr(h=0.1)
```

```python
plt.plot(x,y)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.savefig('h=01.pdf', bbox_inches='tight')
```

```python
plt.plot(x,y)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.savefig('h=1.pdf', bbox_inches='tight')
```
